```python
import pandas as pd

# Mount Google Drive if needed
from google.colab import drive
drive.mount('/content/drive')

# Load CSV (adjust path as needed)
df = pd.read_csv('/content/drive/MyDrive/Au_nanoparticle_dataset.csv')

df.head()
```

```python
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
new_df = df[selected_columns]
```

```python
new_df.head(20)
```

```python
mean = new_df.mean()
std = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean:\n", mean)
print("Standard Deviation:\n", std)
print("Quartiles:\n", quartiles)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 4, figsize=(20, 4))

for i, col in enumerate(new_df.columns):
    axs[i].hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    axs[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)  # diagonal with KDE
g.map_lower(sns.kdeplot, fill=True) # bivariate KDE in lower triangle
```
